```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
a = [1,2,3]
b = [4,5,6,7]
```

This is how the co-ordinate of surface plot looks like:

```python
surface_plot = [(x,y) for x in a for y in b]
print(surface_plot)
```

```python
a,b = np.meshgrid(a,b)

print(a)
print(b)
```

Let's create surface for wider range of data.

```python
a = np.arange(-4,5,0.2)
b = a

a,b = np.meshgrid(a,b)

print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2 + b**2, cmap='rainbow') # Argument Z also be two dimensional.
plt.show()
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2 + b**2, cmap='coolwarm') # Argument Z also be two dimensional.
plt.show()
```

## Contours

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b, a**2 + b**2, cmap='gray')
plt.show()
```

They represent the 2D slicing of a 3D object.  
If we cut that 3D bowls by a straightline, the above surfaces would be the result.

```python
plt.contour(a,b, a**2 + b**2, cmap='rainbow')
plt.title("2D Contour Plot of the 3D bowl")
plt.show()
```
