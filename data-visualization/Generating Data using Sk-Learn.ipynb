## Generating Datasets

Some useful functions to generate datasets available in ```sklearn.datasets```.
- ```make_regression```
- ```make_classification```
- ```make_blobs```
- ```make_moons```

### Generating Regression Data

````python
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
````

````python
X, Y  = make_regression(n_samples=100, n_features=1, n_informative=1, noise=10.8)
````

````python
print(X.shape, Y.shape)
````

````python
plt.scatter(X, Y, marker='.')
plt.show()
````

As we make changes to the noise, the graph is more scattered.

### Generating Classification Data

````python
from sklearn.datasets import make_classification
````

````python
XC, YC = make_classification(n_samples=400, n_features=2, n_informative=2, n_redundant=0)
````

```n_samples``` = No. of samples  
```n_features``` = No. of features  
```n_informative``` = No. of informative features(excluding features like house no. which are not much informative)  
```n_redundant``` = No. of redundant features  

````python
print(XC.shape, YC.shape)
````

````python
plt.scatter(XC[:,0],XC[:,1], c=YC, marker='o')
plt.show()
````

Here, we are seeing two clusters for a class. So, we are reducing it to one.

````python
XC, YC = make_classification(n_samples=400, n_features=2, n_informative=2, n_clusters_per_class=1, n_redundant=0)

print(XC.shape, YC.shape)

plt.scatter(XC[:,0],XC[:,1], c=YC, marker='.')
plt.show()
````

Now, there are only one cluster for each class.

When we want to generate same data, we need to state ```random_state=constant_no```.

````python
XC, YC = make_classification(n_samples=400, n_features=2, n_informative=2, n_clusters_per_class=1, n_redundant = 0, random_state=5)
print(XC.shape, YC.shape)
````

````python
plt.scatter(XC[:,0], XC[:,1], c=YC, marker='.')
plt.show()
````

If we want to increase the no. of classes, the we have to state ```n_classes=no_of_classes```.

````python
XC, YC = make_classification(n_samples=400, n_features=2, n_informative=2, n_redundant=0, random_state=5, n_classes=3, n_clusters_per_class=1)
print(XC.shape, YC.shape)
````

````python
plt.scatter(XC[:,0], XC[:,1], c=YC, marker='.')
plt.show()
````

## Generating Blobs Datasets

````python
from sklearn.datasets import make_blobs
````

````python
XB, YB = make_blobs(n_samples=400, n_features=2, centers=5, random_state=10)
print(XB.shape, YB.shape)
````

````python
plt.scatter(XB[:,0], XB[:,1], c=YB, marker='.')
plt.show()
````

Every clusters depicts separate information.  
It's preferred over classification data because data generated from classification are more close to each other,   
but generated from blobs are far apart from each other.

```make_blobs``` is more used when we are generating datasets for **clusters** as part of **Unsupervised Learning**.

## Generating Moons Datasets

````python
from sklearn.datasets import make_moons
````

````python
XM, YM = make_moons(n_samples=100, shuffle=True, noise=None, random_state=None)
print(XM.shape, YM.shape)
````

````python
plt.scatter(XM[:,0], XM[:,1], c=YM, marker='.')
plt.show()
````

After increasing the noise,

````python
XM, YM = make_moons(n_samples=100, shuffle=True, noise=1.5, random_state=None)
print(XM.shape, YM.shape)
plt.scatter(XM[:,0], XM[:,1], c=YM, marker='o')
plt.show()
````
