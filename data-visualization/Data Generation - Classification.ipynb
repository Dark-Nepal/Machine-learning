Let us generate some examples for Binary classification.  
For this we are using multivariate normal distribution.

We need mean of both data and matrix of co-variance which defines variation.

Here, X1 refers to the size of the fruit whereas,    
X2 refers to the sweetness of the fruit.

Based on this, we will distinguish the fruits.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
u1 = np.array([2, 4])
cov_1 = np.array([[2,-0.8],
                  [-0.8,1]])

u2 = np.array([-1, -4])
cov_2 = np.array([[1,0],
                  [0,2]])
````

u1 is the mean for the first cluster and cov_1 is the co-variance matrix for the first cluster.

````python
X1 = np.random.multivariate_normal(u1, cov_1, 300)
print(X1.shape)

X2 = np.random.multivariate_normal(u2, cov_2, 400)
print(X2.shape)
````

````python
plt.scatter(X1[:,0], X1[:,1], marker='.', label="Apple")
plt.scatter(X2[:,0], X2[:,1], marker='.', label="Orange")
plt.legend()
plt.xlabel("X1 (Size)")
plt.ylabel("X2 (Sweetness)")
plt.show()
````

Let's create a matrix for Y too.

````python
y = np.zeros(700)
y[:300] = 1
print(y.shape)
print(y)
````

````python
X = np.vstack((X1, X2))

print(X.shape)
````

We vertically stack both matrices to form a one.

````python
plt.scatter(X[:,0], X[:, 1], marker='.')
plt.xlabel("X1 (Size)")
plt.ylabel("X2 (Sweetness)")
plt.show()
````

Let's separate the colors for both.

````python
for i in range(700):
    if y[i] == 1:
        plt.scatter(X[i,0], X[i,1], color='red', marker='.')
        plt.xlabel("X1 (Size)")
        plt.ylabel("X2 (Sweetness)")
    else:
        plt.scatter(X[i,0], X[i,1], color='blue', marker='.')
        plt.xlabel("X1 (Size)")
        plt.ylabel("X2 (Sweetness)")
        
plt.show()
````

**Shortcut Method for separating colors**

````python
plt.scatter(X[:,0], X[:,1], marker='.', c=y)
plt.xlabel("X1 (Size)")
plt.ylabel("X2 (Sweetness)")
plt.show()
````

```c=y``` will read the sequence of ```y``` and generate the differnt colors for both.
