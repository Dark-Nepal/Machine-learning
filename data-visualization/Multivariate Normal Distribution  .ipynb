This sort of distribution takes multi parameters.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
mean = np.array([0.0, 0.0])
cov = np.array([[1,0],[0,1]])
````

````python
dist = np.random.multivariate_normal(mean, cov, 500)
print(dist.shape)
````

One column is for X value and another column is for Y value.

````python
plt.scatter(dist[:,0], dist[:,1], marker='.')
plt.show()
````

Here, $\sigma_{xx} = 1,$    
$\sigma_{yy} = 1,$    
$\sigma_{xy} = \sigma_{yx} = 0$  

```x``` and ```y``` are independent of each other.

````python
mean = np.array([0.0, 0.0])
cov = np.array([[1,0.5],[0.5,1]])

dist = np.random.multivariate_normal(mean, cov, 500)
print(dist.shape)

plt.scatter(dist[:,0], dist[:,1], color='red')
plt.show()
````

Changing the covariance matrix , $\sigma_{xy} = \sigma_{yx} = 0.5$, we can observe some sort of positive correlation between them.

Let's make two datasets of multivariate normal distribution where they vary with mean and standard deviation.

````python
mean = np.array([0.0, 0.0])
mean2 = np.array([5.0, 5.0])

cov = np.array([[1,0.9],[0.9,1]])
cov2 = np.array([[1,0.2],[0.2,1]])

dist = np.random.multivariate_normal(mean, cov, 500)
dist2 = np.random.multivariate_normal(mean2, cov2, 500)

plt.scatter(dist[:,0], dist[:,1], color='red', label='Dataset 1', marker='.')
plt.scatter(dist2[:,0], dist2[:,1], color='green', label='Dataset 2', marker='.')
plt.legend()
plt.show()
````
