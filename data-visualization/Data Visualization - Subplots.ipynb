We can invoke ```plt.figure()``` to draw figure in separate figures.

````python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, make_regression, make_blobs, make_moons
````

### Generating Regression Datasets

````python
XR, YR = make_regression(
        n_samples=400,
        n_features=1,
        n_informative=1,
        noise=45,
        random_state=None,
)

print(XR.shape, YR.shape)
````

````python
plt.scatter(XR, YR, color='red', marker='.')
plt.show()
````

### Generating Classification Datasets

````python
XC, YC = make_classification(
        n_samples=400,
        n_features=2,
        n_informative=2,
        n_redundant=0,
        n_classes=3,
        n_clusters_per_class=1,
        random_state=None,
)

print(XC.shape, YC.shape)
````

````python
plt.scatter(XC[:,0], XC[:,1], c=YC, marker='.')
plt.show()
````

### Generating Moons Datasets

````python
XM, YM = make_moons(
        n_samples=100, 
        noise=None, 
        random_state=None
)

print(XM.shape, YM.shape)
````

````python
plt.scatter(XM[:,0], XM[:,1], c=YM, marker='.')
plt.show()
````

### Generating Blobs Datasets

````python
XB, YB =  make_blobs(
        n_samples=400,
        n_features=2,
        centers=5,
        random_state=None
)

print(XB.shape, YB.shape)
````

````python
plt.scatter(XB[:,0], XB[:,1], c=YB, marker='.')
plt.show()
````

### Combining on Subplot

Figure size is the tuple of grids.  
Subplots take (grid and quadrants) as arguments.  
In ```plt.subplot(421)```,    
42 grid means the the subplot will be drawn in one-fourth height, one-half width of whole figure and 1 means in the first quadrant.

````python
plt.figure(figsize=(16, 16))

plt.subplot(221)
plt.title("Regression Datasets")
plt.scatter(XR, YR, color='red', marker='.')

plt.subplot(222)
plt.title("Classification Datasets")
plt.scatter(XC[:,0], XC[:,1], c=YC, marker='.')

plt.subplot(223)
plt.title("Moons Datasets")
plt.scatter(XM[:,0], XM[:,1], c=YM, marker='.')

plt.subplot(224)
plt.title("Blobs Datasets")
plt.scatter(XB[:,0], XB[:,1], c=YB, marker='.')

plt.show()
````

| Moons                                                                                | Blobs                                                                        |
|--------------------------------------------------------------------------------------|------------------------------------------------------------------------------|
| Moons is for non linear data generation which cant be categorized.                   | Blobs is mainly used for making clusters.                                    |
| In Moons Dataset Visualization, we can't classify samples of two different classes.  | In Blobs Dataset Visualization, same class of elements are grouped together. |
