````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.arange(10)
print(x)
````

````python
plt.plot(x)
````

When we pass only one value to the ```plot()``` method, then the value of X will be shifted to x and x acts as indices.

````python
x = x**2
print(x)
````

````python
plt.plot(x)
plt.show()
````

Here, the value of X is shifted to Y , and X acts as indices.

````python
x_one  = x**2.7
plt.plot(x_one)
plt.show()
````

If we want only one graph, then we need to invoke ```show()``` method only after plotting the graph.

````python
plt.plot(x, color='red')
plt.show()
````

We can pass color as keyword arguments.

````python
plt.scatter(x, x**2)
plt.show()
````

Let's add label to the graph.

````python
plt.scatter(x, x**2, color='gold', label='House Price')
plt.legend()
plt.show()
````

If we want to display label we need to invoke ```legend()``` method.

Now, let's add X label and Y label too.

````python
plt.scatter(x, x**2, color='gold', label='House Price')
plt.legend()
plt.xlabel("House Price")
plt.ylabel("Days")
plt.show()
````

Let's change the marker.

````python
plt.scatter(x, x**2, color='red', label='House Price',  marker='^')
plt.legend()
plt.title("House Price Prediction")
plt.xlabel("House Price")
plt.ylabel("Days")
plt.show()
````
