````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
x = np.arange(10)
plt.plot(x, x**2, label="2018")
plt.plot(x, x**3, label="2019")
plt.legend()
plt.xlabel("No of days")
plt.ylabel("Prices")
plt.title("Stock Price Comparision")
plt.show()
````

To distinguish the two different plots, we need to specify label for each and call ```legend()```.

## Bar Chart

````python
numbers = np.random.randint(low=0, high=10, size=5)
indices = np.arange(5)

print(numbers)
print(indices)
````

````python
plt.bar(indices, numbers, color='red')
plt.show()
````

We can specify the width by keyword width.

````python
plt.bar(indices, numbers, width=0.5, color='red')
plt.show()
````

Showing the label for the data:

````python
plt.bar(indices, numbers, width=0.5,color='red', label="Gold Price")
plt.legend()
plt.show()
````

````python
plt.bar(indices*2, numbers, width=0.5, color='red', label="Gold Price")
plt.bar(indices+1, numbers-1, width=0.5, color='green', label="Silver Price")
plt.legend()
plt.show()
````

We can shift the index side by side by increasing/decreasing the value of the indices.

## Pie Chart

````python
labels = ['Algebra', 'Calculus', 'Coding', 'Algorithm']
values = [10, 20, 30, 40]
````

````python
plt.pie(values, labels=labels, radius=1)
plt.title("Prerequisites for Machine Learning")
plt.show()
````

### Highlighting a specific Portion

If we want to highlight a portion, we need to supply a vector for that portion as tuple.

````python
plt.pie(values, labels=labels, radius=1, explode=(0,0,1,0))
plt.title("Prerequisites for Machine Learning")
plt.show()
````

```explode``` parameter will take out the portion we want to highlight.

````python
plt.pie(values, labels=labels, radius=1, explode=(0,0,0,1))
plt.title("Prerequisites for Machine Learning")
plt.show()
````
