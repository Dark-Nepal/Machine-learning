### Standardization

Standardization means to change the raw data into the data with ```mean``` 0 and ```standard deviation``` 1.

For a raw dataset X to standardize,  

$Z = \frac{X - \mu}{\sigma}$
where, 

Z = standardized X  
X = raw dataset  
$\mu$ = mean of raw dataset X  
$\sigma$ = standard deviation of raw dataset X

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
````

````python
os.chdir(r"C:\users\arjun\Desktop\workspace\machine-learning\data-visualization")
print(os.getcwd())
````

````python
X = pd.read_csv('./csv/linearX.csv').values
Y = pd.read_csv('./csv/linearY.csv').values
````

````python
plt.scatter(X,Y,marker='.', label='Raw Data')
plt.legend()
plt.title("Raw Data")
plt.show()
````

Here, the data is raw, i.e. mean isn't ```0``` and standard deviation isn't ```1```.

````python
standardized_X = (X - np.mean(X))/np.std(X)
````

````python
plt.scatter(X,Y,marker='.', label='Raw Data', color='red')
plt.scatter(standardized_X, Y, marker='.', label='Standardized Data', color='green')

plt.legend()
plt.title("Raw v/s Standardized Data")
plt.show()
````

Here, the data is standardized, i.e. mean is ```0``` and standard deviation is ```1```.

### Normalization

The terms normalization and standardization are sometimes used interchangeably,   
but they usually refer to different things.       
Normalization usually means to scale a variable to have a values between 0 and 1.

It basically decreases the spread.  

Actually it's very difficult for the algorithm to work on dataset that has very large standard deviation.  
So, to decrease the spread and combine all the data about mean or any point that will be mean, we normalise the data.

And, most of the real time data sets are in normal distribution only.
