```python
import numpy as np
```

```python
x = np.random.randint(low=0, high=20, size=15)
print(x)
```

```python
print(np.median(x))
```

```python
print(np.average(x))
```

```python
print(np.mean(x))
```

```python
print(np.std(x))
```

```python
print(np.var(x))
```

```python
print(np.histogram(x))
```

First array is for frequency, and second is for the range.

```python
y,z = np.histogram(x)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x, color='red')
plt.title("Graph of random values generated")
plt.show()
```

```python
plt.plot(y, 'o-', color='red')
plt.plot(z, 'o--', color='blue')
plt.title("Relation between Range and Frequency of data")
```
