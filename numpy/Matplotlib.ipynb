It will help to plot the matplotlib in Jupyter Notebook

````python
import numpy as np
import matplotlib.pyplot as plt
````

plt is alias name for matplotlib.pyplot and np for numpy.

````python
x = np.arange(10)
print(x)
````

It will create a numeric range from 0 to n-1.

````python
y = x
````

````python
plt.plot(x,y)
plt.show()
````

````python
plt.plot(x,y,'o')
````

```'o'``` will plot the graph using big dots.

````python
plt.plot(x,y,'o-')
````

```'o-'``` will plot the graph using big dots connected by thin line.

````python
plt.plot(x,y, 'o--')
````

```'o--'``` will plot the graph using big dots connected by thin dotted line.

### Combining two lines in a graph

````python
plt.plot(x,y,'o-')
plt.plot(x,-y,'o--')
````

Matplotlib will separate the lines using different colors.

## Linear Space

For user-defined upper-bound and lower-bound of the data, we use ```linspace``` method.

````python
N = 10
x = np.linspace(0,10,N)
print(x)
y = x
````

0 = lower bound of the data  
10 = upper bound of the data  
N = number of data


````python
plt.plot(x,y, 'o-')
plt.axis('off')
````

 ```plt.axis('off')``` will turn off the display of axes.

## Logarithmic Space

````python
N = 10
x = np.linspace(0, 10, N)
y = np.logspace(0.1, 1, N)
print(y)
plt.plot(x,y, 'o-')
````

## Geometric Space

````python
N = 10
x = np.linspace(0, 10, N)
y = np.geomspace(0.1, 1000, N)
print(y)
plt.plot(x,y, 'o-')
````

For larger values of y, the values are shown in exponential numeric representation.
