````python
import matplotlib.pyplot as plt
import cv2
````

### Read Image

````python
img = cv2.imread('man.jpg')
````

````python
plt.imshow(img)
plt.title('BGR Image')
plt.show()
````

Every color image has 3 color channels of color **Red, Green and Blue** respectively.  
And ```matplotlib``` thinks of the image as BGR channel, i.e. Red and Blue color chhannel are swapped here.

### Switch Between Color Channel

````python
img = cv2.imread('man.jpg')

new_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(new_img)
plt.show()
````
